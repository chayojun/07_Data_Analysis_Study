```python
# 반복문(for)
# 나는 덩어리들에서 처음부터 하나씩 빼가지고 덩어리라고 하겠다. 
덩어리들 = ["덩어리1", "덩어리2", "덩어리3"]
for 덩어리 in 덩어리들:
    # 덩어리에다가 할일들을 쓴다.
    print(덩어리)

print(덩어리) 
## 작동 과정
# 1) 덩어리들에서 맨처음 하나 뺀다. --- 덩어리 = "덩어리1" ---- print(덩어리) "덩어리1"
# 2) 덩어리들에서 두번째 하나 뺀다. --- 덩어리 = "덩어리2" ---- print(덩어리) "덩어리2"
# 3) 덩어리들에서 세번째 하나 뺀다. --- 덩어리 = "덩어리3" ---- print(덩어리) "덩어리3"
```

```python
# 반복문 while
# while 조건:
    # 조건이 True일때 할일들을 쓴다.

# 예시
i = 0
while i < 3:
    print(i)
    i += 1

# i = 0
# i < 3 응 > print(i) 0 > i = 1
# i < 3 응 > print(i) 1 > i = 2
# i < 3 응 > print(i) 2 > i = 3
# i < 3 아니오 > 종료
```
